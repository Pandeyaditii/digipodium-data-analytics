load the data analysis libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall =[17,10,10,10,26,59]
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
```

```python
plt.style.available
```

```python
plt.bar(months, rainfall)
```

```python
plt.style.use('ggplot')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',

)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    wedgeprops={'width': .5 , 'edgecolor': '#6666ff', 'linewidth': 2},
    pctdistance=.7

)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color': 'red'}
            )
plt.show()
```
